```python
import numpy as np
import matplotlib.pyplot as plt

# VBE sweep
VBE = np.linspace(0.5, 0.9, 500)

# Exponential approximation of base current
IB = 0.01 * np.exp((VBE - 0.6) / 0.03)
plt.figure(figsize=(8,4))
plt.plot(VBE, IB, color='blue', linewidth=2)

plt.title("BJT Input Characteristics (Common Emitter)", fontsize=14)
plt.xlabel("$V_{BE}$ (V)", fontsize=12)
plt.ylabel("$I_B$ (mA)", fontsize=12)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# VCE sweep
VCE = np.linspace(0, 10, 500)

# Base currents (µA)
IB_values = [10, 20, 30, 40, 50]
plt.figure(figsize=(8,4))
# Plot characteristic curves
for IB in IB_values:
    IC = (IB / 10) * (1 - np.exp(-VCE / 0.3))  # Approximate model
    plt.plot(VCE, IC, linewidth=2, label=f'$I_B$ = {IB} µA')

# Formatting
plt.title("BJT Output Characteristics (Common Emitter)", fontsize=14)
plt.xlabel("$V_{CE}$ (V)", fontsize=12)
plt.ylabel("$I_C$ (mA)", fontsize=12)
plt.xlim(0, 10)
plt.ylim(0, 6)
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```
